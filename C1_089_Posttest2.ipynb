```python
# Filter yang ada isinya aja yang dihitung
rule_baru = []
for rule in rules:
  # 
  if len(df[(df[rule[0]]==1) & (df[rule[1]]==1)]) >0:
    rule_baru.append(rule)
rule_baru
```

```python
hitung = {} #Ganti dictionary

#bikin dictionary baru dengan nama variabel yang berbeda
for rule in rule_baru:
  banyak_data= len(df[
      (df[rule[0]]==1) & (df[rule[1]]==1)
      ])
  hitung[str(rule)] = banyak_data

hitung
```

```python
# List support
supp = [i/len(df)*100 for i in hitung.values()]
# print(supp)


print("---")
for item,ab in hitung.items():
  print(f"{item} support = {ab/len(df)*100}%") #Untuk menghitung jumlah presentase dari nilai support
```

##Target

```python
# Membuat tabel jarak
# Membuat dataframe dari data baru
dfub = pd.DataFrame({'diagonal'    : [172.25],
                         'height_left' : [103.94],
                         'height_right': [103.68],
                         'margin_low'  : [4.14],
                         'margin_up'   : [3.25],
                         'length'      : [112.48]})
```

```python
# 1. Tambahkan kolom jarak untuk menampilkan jarak dari dataset terhadap target prediksi
```

```python
# Menampilkan data
dfk
```

```python
distances = []
for i in range(len(dfk)):
    row = dfk.iloc[i]
    distance = np.sqrt((row['diagonal'] - dfub['diagonal']) ** 2 +
                       (row['height_left'] - dfub['height_left']) ** 2 +
                       (row['height_right'] - dfub['height_right']) ** 2 +
                       (row['margin_low'] - dfub['margin_low']) ** 2 +
                       (row['margin_up'] - dfub['margin_up']) ** 2 +
                       (row['length'] - dfub['length']) ** 2)
    distances.append(distance)
```

```python
# Memisahkan data dan target
X = dfk.iloc[:, :-1]
y = dfk.iloc[:, -1]
```

```python
# Menghitung jarak dari setiap data point ke target prediksi
distances = np.sqrt(np.sum((X - y.values.reshape(-1, 1)) ** 2, axis=1))
dfk['jarak'] = distances
```

```python
# Menyimpan hasil ke file CSV
dfk.to_csv('deteksi_uang_palsu_jarak.csv', index=False)
```

```python
# Menampilkan hasil file CSV dari penambahan kolom jarak
dfk_jarak = pd.read_csv('deteksi_uang_palsu_jarak.csv')
dfk_jarak
```

```python
# 2. Urutkan data berdasarkan K jarak terdekat (Untuk K nya ambil akar 2 dari banyak data)
```

```python
# Menghitung jarak dari setiap data point ke target prediksi
distances = np.sqrt(np.sum((X - y.values.reshape(-1, 1)) ** 2, axis=1))
```

```python
# Mengambil K jarak terdekat
K = int(np.sqrt(len(dfk)))
indices = np.argsort(distances)[:K]
```

```python
# Mengurutkan data berdasarkan K jarak terdekat
dfk_sorted = dfk.iloc[indices].reset_index(drop=True)
```

```python
# Menyimpan hasil ke file CSV
dfk_sorted.to_csv('deteksi_uang_palsu_sorted.csv', index=False)
```
